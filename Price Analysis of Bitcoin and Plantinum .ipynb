```python
import quandl
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import style
api_key = open('quandlapikey.txt','r').read() 
```

## PRICE ANALYSIS OF PLANTINUM

1.Raw Price printed head and Plotted

```python
df_plat = quandl.get("JOHNMATT/PLAT",authtoken=api_key,start_date="2014-05-14")
df_plat.rename(columns={'New York 9:30':'Plantinum'}, inplace=True)
df_plat.drop({'Hong Kong 8:30', 'Hong Kong 14:00', 'London 09:00'}, inplace=True, axis=1)
df_plat.dropna(inplace = True)
print("Raw Price of Platinum")
print(df_plat.head())
```

```python
df_plat.plot() # Plots graph
plt.ylabel('Raw Price')
plt.title("Raw Stock Price of Plantinum Over 5 years")
```

2. Monthly averaged price and  smoothing of the data applied, printed an plotted

```python
#Resampling Of Plantinum with monthly average 
df_monthly_averaged = df_plat['Plantinum'].resample('M').mean()
df_monthly_averaged.dropna(inplace = True)
print("Monthly Averaged Price of Plantinum")
df_monthly_averaged.head()
```

```python
#Graph of Plantinum averaged monthly
df_monthly_averaged.plot() # Plots graph
plt.ylabel('Price')
plt.title("Monthly Price of Plantinum")
```

3. Monthly averaged price percent smoothing of the data applied, printed an plotted

```python
#Monthly Percent Price change of Platinum 
df_platpercent=df_monthly_averaged.pct_change()
df_platpercent.dropna(inplace = True)
print("Monthly Averaged Price of Plantinum with Percent Change")
df_platpercent.head()
```

```python
#Graph showing Monthly Average Price Percent Change of Plantinum
df_platpercent.plot()  
plt.ylabel('Raw Price')
plt.title("Monthly Average Price Percent Change of Plantinum")
```

4. Maximum, minimum and average price of platinum

```python
df_plat.describe()
```

The Maximum Price for plantinum 5 years ago was $1520

The minimum Price for plantinum 5 years was $770

The average price for plantinum of $1009.79

# STOCK COMPARISON

```python
#RAW PRICE PLOTTED TOGETHER 
fig = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))
df_plat.plot(ax=ax1,label ='Platinum')
df_bitcoin.plot(ax =ax1, label = 'Bitcoin')
plt.legend()
plt.show()
```

```python
#MONTHLY PERCENT AVERAGE PRICE PLOTTED TOGETHER

fig = plt.figure()
ax1 = plt.subplot2grid((1,1), (0,0))
df_monthly_averaged.plot(ax=ax1,label ='Platinum')
df_monthly_averaged_bitcoin.plot(ax =ax1, label = 'Bitcoin')
plt.legend()
plt.show()
```
